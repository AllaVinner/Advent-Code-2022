```python
with open('input.txt' , 'r') as f:
    content = f.read().replace("\r", "")
```

```python
num_col = 9
cols = [[] for _ in range(num_col)]
for (i, line) in enumerate(content.split("\n")):
    if line[0:2] == " 1":
        break
    for col in range(9):
        item = line[4*col+1]
        if item != " ":
           cols[col].append(item)
cols
             
```

```python
def move(num, from_, to_, crates):
    for n in range(num):
        item = crates[from_].pop(0)
        crates[to_] = [item] + crates[to_]
    return crates

def parse_move(s):
    splits = s.split(' ')
    print(splits)
    return int(splits[1]), int(splits[3])-1, int(splits[5])-1
```

```python
for i, l in enumerate(content.split("\n")):
    if i < 10:
        continue
    num, from_, to_ = parse_move(l)
    cols = move(num, from_, to_, cols)
    
```

```python
[c[0] for c in cols]
```

```python
def move2(num, from_, to_, crates):
    crates[to_] = crates[from_][0:num] + crates[to_]
    for n in range(num):
        item = crates[from_].pop(0)
    return crates
```
