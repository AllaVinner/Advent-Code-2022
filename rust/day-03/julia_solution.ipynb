```python
with open('input.txt' , 'r') as f:
    content = f.read()
content
```

```python
bags = content.split("\n")
bags
```

```python
sum = 0
for bag in bags:
    bagsize = len(bag)
    compsize = bagsize // 2 
    comp1 = bag[0:compsize] 
    comp2 = bag[compsize:]
    alphabet = {}
    for item1 in comp1:
        alphabet[item1] = 1
    for item2 in comp2:
        dot = alphabet.get(item2, 0)
        if dot == 1:
            if item2.isupper():
                sum = sum + ord(item2) - 65 + 27
            else:
                sum = sum + ord(item2) - 96
            break
        
        
sum
```
